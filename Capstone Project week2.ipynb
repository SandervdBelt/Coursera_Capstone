```python
#Now let's filter the extensive JSON files to what I actually need. For that, I'll need this little formula

def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
spots_Hasselt= Hasselt['response']['groups'][0]['items']
spots_Hasselt2=json_normalize(spots_Hasselt)
    
# filter columns
filtered_columns = ['venue.location.city', 'venue.name', 'venue.categories']
spots_Hasselt3=spots_Hasselt2.loc[:, filtered_columns]

# filter the category for each row
spots_Hasselt3['venue.categories'] = spots_Hasselt3.apply(get_category_type, axis=1) #here the formula is being used

spots_Hasselt3
```
